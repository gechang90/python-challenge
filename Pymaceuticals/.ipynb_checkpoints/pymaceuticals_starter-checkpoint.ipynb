```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_data = pd.read_csv(mouse_drug_data_to_load)
df_mouse = pd.DataFrame(mouse_drug_data)
clinical_trial_data = pd.read_csv(clinical_trial_data_to_load)
df_clinical = pd.DataFrame(clinical_trial_data)

# Combine the data into a single dataset
df_mouse_clinical = pd.merge(df_mouse,df_clinical)

# Display the data table for preview
df_mouse_clinical.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
ob_mean_tumor_vol = df_mouse_clinical.groupby(['Drug','Timepoint'])['Tumor Volume (mm3)'].mean()
df_mean_tumor = ob_mean_tumor_vol.to_frame().reset_index()
# Preview DataFrame
df_mean_tumor.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
ob_stderror__tumor_vol = df_mouse_clinical.groupby(['Drug','Timepoint'])['Tumor Volume (mm3)'].sem()
# Convert to DataFrame
df_stderror = ob_stderror__tumor_vol.to_frame().reset_index()
# Preview DataFrame

df_stderror.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
df_reformat_mean_tumor = df_mean_tumor.pivot(index = 'Timepoint',columns='Drug')


# Preview that Reformatting worked
df_reformat_mean_tumor.head()
```

```python
# Generate the Plot (with Error Bars)
x_axis = np.arange(0,50,5)
capomulin = df_reformat_mean_tumor['Tumor Volume (mm3)']['Capomulin']
Infubinol = df_reformat_mean_tumor['Tumor Volume (mm3)']['Infubinol']
Ketapril = df_reformat_mean_tumor['Tumor Volume (mm3)']['Ketapril']
Placebo = df_reformat_mean_tumor['Tumor Volume (mm3)']['Placebo']

capomulin_plot, = plt.plot(x_axis, capomulin, marker ='o', color='red', label="Capomulin")
Infubinol_plot, = plt.plot(x_axis, Infubinol, marker ='^', color='blue', label="Infubinol")
Ketapril_plot, = plt.plot(x_axis, Ketapril, marker ='s', color='green', label="Ketapril")
Placebo_plot, = plt.plot(x_axis, Placebo, marker ='d', color='black', label="Placebo")
plt.legend(handles=[capomulin_plot,Infubinol_plot,Ketapril_plot,Placebo_plot ], loc="best")
plt.ylabel('Tumor Volume(mm3)')
plt.xlabel("Time(Days)")
plt.grid(True)
# Save the Figure
plt.savefig("Tumor Response to Treatment.png")
```

```python
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
mean_met_data = df_mouse_clinical.groupby(['Drug','Timepoint'])['Metastatic Sites'].mean()

# Convert to DataFrame
df_reset1 = mean_met_data.to_frame().reset_index()
df_met_data = mean_met_data.to_frame()
# Preview DataFrame
df_met_data.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
std_error_data = df_mouse_clinical.groupby(['Drug','Timepoint'])['Metastatic Sites'].sem()

# Convert to DataFrame
df_std_error = std_error_data.to_frame()

# Preview DataFrame
df_std_error.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
df_reformat_met = df_reset1.pivot(index ="Timepoint",columns='Drug')

# Preview that Reformatting worked
df_reformat_met.head()
```

```python
# Generate the Plot (with Error Bars)
capomulin_1 = df_reformat_met['Metastatic Sites']['Capomulin']
Infubinol_1 = df_reformat_met['Metastatic Sites']['Infubinol']
Ketapril_1 = df_reformat_met['Metastatic Sites']['Ketapril']
Placebo_1 = df_reformat_met['Metastatic Sites']['Placebo']
capomulin_plot_1, = plt.plot(x_axis, capomulin_1, marker ='o', color='red', label="Capomulin")
Infubinol_plot_1, = plt.plot(x_axis, Infubinol_1, marker ='^', color='blue', label="Infubinol")
Ketapril_plot_1, = plt.plot(x_axis, Ketapril_1, marker ='s', color='green', label="Ketapril")
Placebo_plot_1, = plt.plot(x_axis, Placebo_1, marker ='d', color='black', label="Placebo")
plt.legend(handles=[capomulin_plot_1,Infubinol_plot_1,Ketapril_plot_1,Placebo_plot_1 ], loc="best")
plt.ylabel('Metastatic Sites')
plt.xlabel("Time(Days)")
plt.grid(True)
# Save the Figure
plt.savefig("Metastatic Response to Treatment.png")
# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
ob_survival_rates = df_mouse_clinical.groupby(['Drug','Timepoint'])['Mouse ID'].count()
# Convert to DataFrame
df_survival_rates = ob_survival_rates.to_frame().reset_index()
# Preview DataFrame
df_survival_rates.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
df_survival_reformat = df_survival_rates.pivot(index = 'Timepoint',columns = 'Drug')

# Preview the Data Frame
df_survival_reformat.head()
```

```python
# Generate the Plot (Accounting for percentages)
capomulin_2 = df_survival_reformat['Mouse ID']['Capomulin']
Infubinol_2 =df_survival_reformat['Mouse ID']['Infubinol']
Ketapril_2= df_survival_reformat['Mouse ID']['Ketapril']
Placebo_2 = df_survival_reformat['Mouse ID']['Placebo']
capomulin_plot_2, = plt.plot(x_axis, capomulin_2, marker ='o', color='red', label="Capomulin")
Infubinol_plot_2, = plt.plot(x_axis, Infubinol_2, marker ='^', color='blue', label="Infubinol")
Ketapril_plot_2, = plt.plot(x_axis, Ketapril_2, marker ='s', color='green', label="Ketapril")
Placebo_plot_2, = plt.plot(x_axis, Placebo_2, marker ='d', color='black', label="Placebo")
plt.legend(handles=[capomulin_plot_1,Infubinol_plot_2,Ketapril_plot_2,Placebo_plot_2 ], loc="best")
plt.ylabel("Survival Rates (%)")
plt.xlabel("Time(Days)")
plt.grid(True)

# Save the Figure
plt.savefig("Survival Rates.png")
# Show the Figure
plt.show()
```
